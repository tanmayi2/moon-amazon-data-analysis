```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("inventory.csv")
print(data.head(10))
```

```python
# bar graph of inventory stock
fig = plt.figure(figsize=(10,5))
plt.title("Inventory Stock")
sns.barplot(
    data=data,
    x='available',
    y='sku',
    palette='dark',
    width=0.8,
    alpha=0.6,
    order=data.sort_values('available').sku
)
plt.show()
```
